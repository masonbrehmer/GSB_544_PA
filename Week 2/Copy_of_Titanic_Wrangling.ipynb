```python
# Import the titanic dataset here
import numpy as np
import pandas as pd

data_dir = "https://dlsun.github.io/pods/data/"
df_titanic = pd.read_csv(data_dir + "titanic.csv")
df_titanic
```

## 1. Filter the data to include passengers only. Calculate the joint distribution (cross-tab) between a passenger's class and where they embarked.

```python
# Code here
#Defining the 1st, 2nd, and 3rd class passengers as passenger.
df_titanic["passenger"] = df_titanic["class"].map({
    "1st": "passenger",
    "2nd": "passenger",
    "3rd": "passenger",
    "victualling crew": "crew",
    "engineering crew": "crew",
    "deck crew": "crew"
})
# Keeping only the passengers in the dataframe.
df_titanic = df_titanic[df_titanic["passenger"]=="passenger"]
# Creating a crosstab between passenger class and where they embarked.
pd.crosstab(df_titanic["class"], df_titanic["embarked"])
```

## 2. Using the joint distribution that calculated above, calculate the following:

* the conditional distribution of their class given where they embarked
* the conditional distribution of where they embarked given their class

Use the conditional distributions that you calculate to answer the following quesitons:

* What proportion of 3rd class passengers embarked at Southampton?
* What proportion of Southampton passengers were in 3rd class?

```python
# Code here
# Calculating the conditional distribution of their class given where they embarked.
pd.crosstab(df_titanic["class"], df_titanic["embarked"], normalize="columns")
# 54.05% of Southampton passengers were in 3rd class.
```

```python
# Calculating the conditional distribution of where they embarked given their class.
Conditional_distribution = pd.crosstab(df_titanic["class"], df_titanic["embarked"], normalize="index")
Conditional_distribution
# 69.67% of 3rd class passengers embarked at Southampton.
```

```python
import matplotlib.pyplot as plt

# Calculate the conditional distribution of class given embarked
class_given_embarked = pd.crosstab(df_titanic["embarked"], df_titanic["class"], normalize='index')

# Create a stacked bar chart
ax = class_given_embarked.plot(kind='bar', stacked=False, figsize=(10, 6))

plt.title('Conditional Distribution of Class Given Embarked Location')
plt.xlabel('Embarkment Location')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.legend(title='Class')
plt.grid(axis='y')

# Add percentages to the bars (optional, can make the plot cluttered)
# for container in ax.containers:
#     ax.bar_label(container, fmt='%.2f', label_type='center')

plt.show()
```
